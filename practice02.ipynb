```python
df = pd.read_csv('sample.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (6,6))
plt.scatter(df.Balls,df.Runs)
plt.title('Runs vs Balls - Distribution')
plt.xlabel('Balls')
plt.ylabel('Runs')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.plot(df.Balls,df.Runs)
plt.title('Runs vs Balls - Distribution')
plt.xlabel('Balls')
plt.ylabel('Runs')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.bar(df.Balls,df.Runs)
plt.title('Runs vs Balls - Distribution')
plt.xlabel('Balls')
plt.ylabel('Runs')
plt.show()
```
